##### X, Y 분할

```python
# 피처와 레이블 분리
X = df.drop('stroke', axis=1)
Y = df['stroke']
```

##### 학습 데이터와 평가 데이터 분할

```python
# 학습 데이터와 평가 데이터 분할
from sklearn.model_selection import train_test_split

Train_X, Test_X, Train_Y, Test_Y = train_test_split(X, Y)
```

##### 결측치 / 비율 / 상관관계 확인

```python
# 결측치 확인 : 각 컬럼별로 결측치 개수 확인
Train_X.isnull().sum(axis=0)

# 레코드가 5000개 이상이다보니 결측이 있는 레코드만 삭제해도 무방할 것 같지만
# 새로 들어올 데이터에 결측치가 있을거라고 가정(도메인 지식)하고 결측치를 대체하겠음
```

```python
# 열별 결측치 비율 확인
Train_X.isnull().sum(axis=0) / len(Train_X)
```

```python
# 상관계수 평균 확인
Train_X.corr().sum() / len(Train_X.columns) # 열 별 관련성 : 낮은 관계성 == 대체값으로 활용 가능
```

##### bmi 컬럼 결측치 대체

```python
# 대표값 이용해서 결측치 대체
from sklearn.impute import SimpleImputer

# 빈 모델(객체 인스턴스) 생성
SI = SimpleImputer(strategy = 'mean')

# 학습
SI.fit(Train_X)
SI.transform(Train_X) # arr 형태로 결과 반환
```

```python
# sklearn instance의 출력은 ndarray이므로 다시 DataFrame으로 바꿔줌
Train_X = pd.DataFrame(SI.transform(Train_X), columns = Train_X.columns)
```

```python
# Test_X는 fit을 진행하지 않는다!
Test_X = pd.DataFrame(SI.transform(Test_X), columns=Test_X.columns)
```
